### Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. Здесь будут описаны ключевые правила оформления графиков для **всех** домашних заданий.

У matplotlib, конечно же, есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторой величины;
- scatter — график, показывающий взаимосвязь двух величин;

В этом задании вы попробуете построить один из них. Не забывайте про базовые принципы построения приличных графиков:
- оси должны быть подписаны, причём не слишком мелко;
- у графика должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах);
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать;
- если распределение на гистограмме имеет тяжёлые хвосты, лучше использовать логарифмическую шкалу.

### 2. Распределение студентов по курсам.

Теперь вам нужно распределить студентов по осенним курсам по выбору, учитывая их предпочтения.

Алгоритм распределения студентов по курсам:
1. По умолчанию на каждой дисциплине по выбору у 3 и 4 курсов может учиться 1 группа (до 30 студентов). Исключения описаны ниже. На blended-дисциплинах для четверокурсников количество мест не ограничено.
2. Проводится первая волна отбора. Для каждой дисциплины формируется список тех, кто указал её первым приоритетом (если студент должен выбрать два курса по выбору, то для него дисциплины, которые он указал первым и вторым приоритетом, рассматриваются как дисциплины первого приоритета). Если желающих больше, чем мест, то выбирается топ по перцентилю рейтинга.
3. На дисциплинах, где остались места после первой волны, формируются списки тех, кто выбрал их вторым приоритетом, и места заполняются лучшими по перцентили рейтинга студентами. После этого проводится такая же процедура для дисциплин третьего приоритета.
4. Если студент не попал на необходимое количество курсов по итогам трёх волн, с ним связывается учебный офис и решает вопрос в индивидуальном порядке.

Обращаем ваше внимание на следующие детали:

- По умолчанию студент выбирает один осенний и один весенний курс по выбору, а также четверокурсники выбирают один blended-курс. Студенты групп 17-го года специализаций МОП и ТИ выбирают по 2 осенних и 2 весенних курса по выбору, также студенты групп 18' специализации МИ выбирают 2 осенних курса.

- Студенты специализации МОП не могут выбрать весенним курсом по выбору Машинное обучение 2.

- Blended-курсы не трогайте, по ним не надо распределять, на другие курсы они никак не влияют.

- Постарайтесь воздержаться от использования циклов там, где это возможно.

#### 2. [2 балла] Распределите студентов в соответствии с первым приоритетом

```python
df = df.sort_values(by='percentile')
df['is_first_place'] = np.nan
df['is_last_place'] = np.nan
df_fall = pd.DataFrame(columns=['id', 'course_1'])
```

```python
# функция для проведения одной волны распределения одного семестра
def distribute(df, sub_course, sub_group, season, df_, df_prev, wave):
    
    # таблица, columns которой есть курсы, а содержимое – ids студентов
    distribution = pd.DataFrame(index=np.arange(df.shape[0]))
    
    for course in sub_course:
        
        limit = 60 if course in {'Statistical Learning Theory',
                                 'Высокопроизводительные вычисления',
                                 'Анализ неструктурированных данных',
                                 'Обучение с подкреплением',
                                 'Анализ данных в бизнесе'} else 30
        
        # учитываем то, что в предыдущих волнах часть мест на курсе была занята
        if wave != 1:
            taken = df_prev[df_prev['course_1'] == course].shape[0] + df_prev[df_prev['course_2'] == course].shape[0]
            if taken >= limit:
                continue
            limit -= taken
        
        
        # выбираем студентов, подходящих для курса и данной волны
        if wave == 1:
            group = sub_group[(sub_group[f'{season}_1'] == course)
                              | (sub_group[f'{season}_2'] == course) & (sub_group[f'{season}_1'] != course)
                              & (sub_group[f'n_{season}'] == 2)]
                              
        elif wave == 2:
            group = sub_group[(sub_group[f'{season}_2'] == course)
                              & (sub_group[f'n_{season}'] == 1)
                              & (sub_group['course_1'].isna())
                              | (sub_group[f'n_{season}'] == 2)
                              & ((sub_group['course_1'].isna())
                                 | ((sub_group['course_2'].isna()) & (sub_group['course_1'] != course)))
                              & (sub_group[f'{season}_3'] == course)]
                              
        else:
            group = sub_group[(sub_group[f'{season}_3'] == course)
                              & (sub_group[f'n_{season}'] == 1)
                              & (sub_group['course_1'].isna())]
        
        group = group[:min(group.shape[0], limit)]
        distribution[course] = group['id'].reset_index(drop=True)
        
        # помечаем, является ли студент лучшим/худшим на курсе (только для первой волны осени)
        if wave == 1 and season == 'fall' and not group.empty:
            first, last = group.id.iloc[0], group.id.iloc[group.shape[0] - 1]
            df.loc[df['id'] == first, 'is_first_place'] = True
            df.loc[df['id'] == last, 'is_last_place'] = True
    
    # переворачиваем табличку так, чтобы каждому айди соответствовал назначенный курс
    distribution = distribution.stack().reset_index(level=1, name='id').rename(columns={
        'level_1': 'course_1'
    })
    distribution = distribution[['id', 'course_1']]
    df_ = pd.concat((df_, distribution))
    # возвращаем таблицу, формируемую по мере распределения студентов по курсам
    return df_
```

```python
# функция для переворачивания табличек и обновления значений
def helper(df, season, curr, prev, courses_17, courses_18, common, start=1, end=4):
    # списки курсов для студентов 17, 18 годов и общие курсы
    only_17 = courses_17[~courses_17.value.isin(common)].value.unique()
    only_18 = courses_18[~courses_18.value.isin(common)].value.unique()
    common = pd.merge(courses_17.value, courses_18.value, how="inner").value.unique()
    
    # для каждой волны
    for i in range(start, end):
        # находим студентов, и, распределяя их по курсам, заполняем табличку
        df_17 = df_17 = df[df['17_group'] != 0]
        curr = distribute(df, only_17, df_17, season, curr, prev, i)
        
        df_18 = df[df['18_group'] != 0]
        curr = distribute(df, only_18, df_18, season, curr, prev, i)
        
        curr = distribute(df, common, df, season, curr, prev, i)
        
        # пришиваем результаты этой волны к результатам предыдущей (если таковая имеется)
        if i > 1:
            curr = pd.concat((prev, curr)).reset_index(drop=True)
        
        # помечаем айдишники, которые дублируются (они уже есть в таблице и, вероятно, получили course2)
        curr['duplicated'] = curr.id.duplicated()
        m = (3, 2)[i == 1]

        # если это первая волна, то все повторы очевидно есть course2,
        # иначе создаем колонку course3, чтобы не потерять course2
        duplicated = curr[curr['duplicated']][['id', 'course_1']].rename(columns={'course_1': f'course_{m}'})
        curr = curr.merge(duplicated, on='id', how='left').drop_duplicates(subset='id')
        curr = curr.drop(columns='duplicated')
        
        # содержимым course3 заполняем пропуски в колонке course2
        if i > 1:
            curr['course_2'] = curr['course_2'].fillna(curr['course_3'])
            curr = curr.drop(columns='course_3')

        curr = curr.merge(df[['id', f'n_{season}']], on='id', how='left')
        
        # если курсов по выбору должно быть 2, но это не последняя волна пропуск мы не заполняем
        curr[f'n_{season}'] = curr[f'n_{season}'].replace({1: '-',
                                                           2: '???' if i == 3 else np.nan})
        curr['course_2'] = curr['course_2'].fillna(curr[f'n_{season}'])
        curr = curr.drop(f'n_{season}', axis=1)
        
        # обновляем результаты распределния в общей таблице
        if i > 1:
            df = df.drop(columns=['course_1', 'course_2'])
        df = df.merge(curr, on='id', how='left')

        prev = curr
    
    return df, prev
```
